```python
mae = mean_absolute_error(y,y_pred)
mse = mean_squared_error(y,y_pred)
rmse = np.sqrt(mse)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE :", rmse)
```

Interpretasi:

MAE = rata-rata kesalahan absolut prediksi

MSE = rata-rata kuadrat kesalahan

RMSE = akar dari MSE (lebih mudah diinterpretasi)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("data salary.csv")

X = df[['YearsExperience']]
y = df['Salary']
```

```python
model = LinearRegression()
model.fit(X,y)

y_pred = model.predict(X)

plt.scatter(X,y)
plt.plot(X,y_pred)

plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("Scatter Plot Regresi Linear")

plt.show()
```

Penjelasan:

Titik = data asli

Garis lurus = model regresi linear

```python
poly2 = PolynomialFeatures(degree=2)
X_poly2 = poly2.fit_transform(X)

model2 = LinearRegression()
model2.fit(X_poly2,y)

y_pred2 = model2.predict(X_poly2)

plt.scatter(X,y)
plt.plot(X,y_pred2)

plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("Scatter Plot Polynomial Regression Orde 2")

plt.show()
```

Grafik mulai sedikit melengkung mengikuti data.

```python
poly3 = PolynomialFeatures(degree=3)
X_poly3 = poly3.fit_transform(X)

model3 = LinearRegression()
model3.fit(X_poly3,y)

y_pred3 = model3.predict(X_poly3)

plt.scatter(X,y)
plt.plot(X,y_pred3)

plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("Scatter Plot Polynomial Regression Orde 3")

plt.show()
```

Kurva menjadi lebih fleksibel mengikuti pola data.

```python
poly4 = PolynomialFeatures(degree=4)
X_poly4 = poly4.fit_transform(X)

model4 = LinearRegression()
model4.fit(X_poly4,y)

y_pred4 = model4.predict(X_poly4)

plt.scatter(X,y)
plt.plot(X,y_pred4)

plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("Scatter Plot Polynomial Regression Orde 4")

plt.show()
```

Kurva semakin mengikuti titik data.

```python
poly5 = PolynomialFeatures(degree=5)
X_poly5 = poly5.fit_transform(X)

model5 = LinearRegression()
model5.fit(X_poly5,y)

y_pred5 = model5.predict(X_poly5)

plt.scatter(X,y)
plt.plot(X,y_pred5)

plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("Scatter Plot Polynomial Regression Orde 5")

plt.show()
```

Model mulai sangat melengkung dan berpotensi overfitting.

Kesimpulan Grafik

1️⃣ Regresi Linear

Hubungan terlihat linear antara pengalaman dan gaji.

2️⃣ Polynomial Orde 2–3

Kurva mengikuti pola data lebih baik.

3️⃣ Polynomial Orde 4–5

Model terlalu mengikuti titik data → overfitting
